##Solution

```
import threading
import time
from random import randrange

class Person:
    def __init__(self, age, salary, taxes):
        self.age = age
        self.salary = salary
        self.taxes = taxes

class MainThread(threading.Thread):

    def __init__(self, name, my_list, lock):
        threading.Thread.__init__(self)
        self.my_list = my_list
        self.name = name
        self.lock = lock

    def run(self):
        lock.acquire()
        for person in self.my_list:
            if person.age in personDict:
                value = personDict.get(person.age)
                time.sleep(randrange(1, 50)/1000)
                personDict[person.age] = value + person.salary - person.taxes
            else:
                 personDict[person.age] = person.salary - person.taxes
        lock.release()


lock = threading.Lock()
personDict = {}
threads = []
# generate list with persons

for salary in range(0, 10):
    personList = []
    for age in range(20, 30):
        personList.append(Person(age, salary, salary * 0.15))

    thread = MainThread("thread_" + str(salary), personList, lock)
    threads.append(thread)

for t in threads:
    t.start()
for t in threads:
    t.join()
print(personDict.values())
```
