#Solution

```
from pyspark.sql import SparkSession
from pyspark.sql.functions import split, explode, desc, length, col

spark = SparkSession.builder.appName("Exercise").getOrCreate()
# Read file by lines
linesDF = spark.read.text("Moby-Dick.txt").select('value')
linesDF.show(truncate=False)
# Split words in line by ' '
linesDF = linesDF.select(split(linesDF.value, ' ').alias('split'))
linesDF.show(truncate=False)
# Add own row to every word
wordsDF = linesDF.select(explode(linesDF.split).alias('words'))
wordsDF.show()
# Filter empty lines
# wordsDF = wordsDF.filter('words != \'\'')
# or with function length
wordsDF = wordsDF.filter(length(col("words")) > 0)
wordsDF.show()
# Group according to words and count same words
wordCountDF = wordsDF.groupBy("words").count()
wordCountDF.show()
# Order words by occurencies
wordCountDF = wordCountDF.orderBy(desc("count"))
wordCountDF.show()
# Print the number of all the word types
wordsSumDF = wordCountDF.agg({'words':'count'})
wordsSumDF.show()
# OR
print("Count word types: "+str(wordCountDF.count()))
# Print the number of all the words occurences
wordsSumDF = wordCountDF.agg({'count':'sum'})
wordsSumDF.show()

spark.stop()
```
